```python
DATADIR = "yang91"
DATADIR2 = "yang91"
#DATADIR2 = "DIV2K_train_HR"
```

```python
import os
import numpy as np
import cv2
```

```python
scale = 2
req_size = 20
inp_size = int(req_size/scale)
stride = 15

X=[]
labels=[]

for filera in os.listdir(DATADIR):
  readpath = os.path.join(DATADIR,filera)
  img = cv2.imread(readpath)
  img = np.asarray(img)
  shapes = img.shape
  for i in range(0, shapes[0]-req_size+1, stride):
    for j in range(0, shapes[1]-req_size+1,stride):
      subimage_HR = img[i:i+req_size, j:j+req_size]
      #cv2_imshow(subimage_HR)
      subimage_LR = cv2.resize(subimage_HR,(inp_size,inp_size))
      X.append(subimage_LR)
      labels.append(subimage_HR)

counta=0
for filera in os.listdir(DATADIR2):
  counta=counta+1
  if(counta==300):
    break
  readpath = os.path.join(DATADIR2,filera)
  img = cv2.imread(readpath)
  img = np.asarray(img)
  shapes = img.shape
  for i in range(0, shapes[0]-req_size+1, stride):
    for j in range(0, shapes[1]-req_size+1,stride):
      subimage_HR = img[i:i+req_size, j:j+req_size]
      #cv2_imshow(subimage_HR)
      subimage_LR = cv2.resize(subimage_HR,(inp_size,inp_size))
      X.append(subimage_LR)
      labels.append(subimage_HR)
```

```python
X = np.asarray(X)
labels = np.asarray(labels)
print(labels.shape)
print(X.shape)
```

```python
import tensorflow as tf
import keras
import numpy as np
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Dense, Conv2D, Activation, Input,Conv2DTranspose
from tensorflow.keras.layers import LeakyReLU
from keras.callbacks import ModelCheckpoint
from keras.optimizers import Adam, SGD
```

```python
alpha = 0.85
def basic_mse(y_true, y_pred):
  pred = tf.image.sobel_edges(y_pred)
  true = tf.image.sobel_edges(y_true)
  lossmse = tf.reduce_mean(tf.square(y_pred-y_true))
  lossedge = tf.reduce_mean(tf.square(pred-true))
  return alpha*lossmse + (1-alpha)*lossedge
```

```python
inputlayer = tf.keras.layers.Input(shape=(None,None,3))
C1 = Conv2D(56, (5,5), strides=(1,1), padding="same",name='CONV1')(inputlayer)
A1 = Activation('relu', name='act1')(C1)

#the layer to reduce feature maps from d to s
C2 = Conv2D(12, (1,1),padding="same",name='CONV2')(A1)
A2 = Activation('relu', name='act2')(C2)

C3 = Conv2D(12, (3,3),padding="same",name='CONV3')(A2)
A3 = Activation('relu', name='act3')(C3)

C4 = Conv2D(12, (3,3),padding="same",name='CONV4')(A3)
A4 = Activation('relu', name='act4')(C4)
Sum1  = tf.keras.layers.add([A3,A4])
C5 = Conv2D(12, (3,3),padding="same",name='CONV5')(Sum1)
A5 = Activation('relu', name='act5')(C5)

C6 = Conv2D(12, (3,3),padding="same",name='CONV6')(A5)
A6 = Activation('relu', name='act6')(C6)

C7 = Conv2D(56, (1,1),padding="same",name='CONV7')(A6)
A7 = Activation('relu', name='act7')(C7)


#final expanding layer of the model 3 channels for RGB
y = Conv2DTranspose(3, (9,9), strides = (scale,scale), padding="same",activation=None,name='UPSCALING')(A7)

model = Model(inputs=inputlayer , outputs = y)
model.summary()
```

```python
tf.keras.utils.plot_model(model, 'my_first_model.png')
```

```python
from keras.callbacks import ModelCheckpoint
from keras.optimizers import Adam, SGD
```

```python
model.compile(loss='mse',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
checkpoint = ModelCheckpoint('best_model_improved.h5',  # model filename
                             monitor='val_loss', # quantity to monitor
                             verbose=0, # verbosity - 0 or 1
                             save_best_only= True, # The latest best model will not be overwritten
                             mode='auto') # The decision to overwrite model is made 
                                          # automatically depending on the quantity to monitor
```

```python
history = model.fit(X,labels,batch_size=32,shuffle=True,validation_split=0.01 ,epochs=20,callbacks=[checkpoint],
                    verbose=1 )
```
